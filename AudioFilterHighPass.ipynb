```python
window_size = 256
hop_size = window_size // 2
total_time = len(input_signal) / sample_rate
fft_bin_width = sample_rate / window_size
```

```python
# Full "filter" algorithm
stft_output = stft(input_signal, sample_rate, window_size, hop_size)
lowpass = high_pass_filter(2500, window_size, fft_bin_width)
filtered_result = [original * lowpass for original in stft_output]
resynth = istft(filtered_result, sample_rate, window_size, hop_size, total_time)
```

```python
plot_large(resynth)
```

```python
plot_large(input_signal)
```

```python
outfile = "audio_files/highpassout.wav"
output_file = audio.Sndfile(outfile, 'w', audio.Format('wav'), 1, sample_rate)
output_file.write_frames(resynth)
output_file.close()
```
