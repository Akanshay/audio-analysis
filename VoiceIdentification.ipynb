# FFT Approach

```python
import scikits.audiolab as audio
import scipy
from __future__ import division

def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
caleb_c3_file = "audio_files/voice_id/caleb_c3.wav"
whitney_c3_file = "audio_files/voice_id/whitney_c3.wav"
caleb_a4_file = "audio_files/voice_id/caleb_a4.wav"

# Input
(caleb_c3, caleb_c3_sample_rate, caleb_c3_bits) = audio.wavread(caleb_c3_file)
(whitney_c3, whitney_c3_sample_rate, whitney_c3_bits) = audio.wavread(whitney_c3_file)
(caleb_a4, caleb_a4_sample_rate, caleb_c3_bits) = audio.wavread(caleb_a4_file)
```

```python
caleb_c3_fft = numpy.fft.rfft(caleb_c3[0:44100])
whitney_c3_fft = numpy.fft.rfft(whitney_c3[0:44100])
caleb_a4_fft = numpy.fft.rfft(caleb_a4[0:44100])

caleb_c3_fft_mag = [sqrt(i.real**2 + i.imag**2)/len(caleb_c3_fft) for i in caleb_c3_fft]
whitney_c3_fft_mag = [sqrt(i.real**2 + i.imag**2)/len(whitney_c3_fft) for i in whitney_c3_fft]
caleb_a4_fft_mag = [sqrt(i.real**2 + i.imag**2)/len(caleb_a4_fft) for i in caleb_a4_fft]
```

# Cross-correlation Approach

```python
import scipy.signal
```

```python
saved_response_f = "audio_files/voice_id/caleb_c3.wav"
attempt1_f = "audio_files/voice_id/caleb_c3.wav"
attempt2_f = "audio_files/voice_id/caleb_c3_2.wav"
attempt3_f = "audio_files/voice_id/caleb_c3_3.wav"

(saved_response, _, _) = audio.wavread(saved_response_f)
(attempt1, _, _) = audio.wavread(attempt1_f)
(attempt2, _, _) = audio.wavread(attempt2_f)
(attempt3, _, _) = audio.wavread(attempt3_f)
```

```python
c1 = scipy.signal.correlate(saved_response, attempt1)
```
