```python
import scikits.audiolab as audio
infile = "audio_files/together_1channel_4sec.wav"
```

```python
def plot_large(*args, **kwargs):
    fig = pylab.figure()
    fig.set_size_inches(14,8)
    plot(*args, **kwargs)
```

```python
(input_signal, sampling_rate, bits) = audio.wavread(infile)
```

```python
plot_large(input_signal)
```

```python
fft_data = numpy.fft.fft(input_signal)
```

```python
print "Sampling rate:", sampling_rate
```

```python
signal_len = len(input_signal)
sample_spacing = 1./sampling_rate # 1 second divided by samples per second
freqs = numpy.fft.fftfreq(signal_len, d=sample_spacing)
```

```python
plot_large(fft_data)
```

```python
plot_large(freqs[0:signal_len/2], fft_data[0:signal_len/2])
```

```python
fft_data2 = numpy.fft.fft(input_signal, n=44100)
```

```python
print "Nyquist frequency:", len(fft_data2)/2
```

```python
# FFT output is complex numbers; where in a+bi, (a,b) angle is the phase, sqrt(a**2+b**2) is magnitude
fft_data2_magnitude = [sqrt(real(comp)**2 + imag(comp)**2) for comp in fft_data2]
```

```python
plot_large(fft_data2_magnitude)
```
