```python
def plot_large(*args, **kwargs):
    fig = pylab.figure()
    fig.set_size_inches(14,8)
    plot(*args, **kwargs)
```

```python
import scipy
import scikits.audiolab as audio
from __future__ import division
infile = "audio_files/ohm_scale.wav"
```

```python
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
                     for i in range(0, len(x)-framesamp, hopsamp)])
    return X

def istft(X, fs, T, hop):
    x = scipy.zeros(T*fs)
    framesamp = X.shape[1]
    hopsamp = int(hop*fs)
    for n,i in enumerate(range(0, len(x)-framesamp, hopsamp)):
        x[i:i+framesamp] += scipy.real(scipy.ifft(X[n]))
    return x
```

```python
(input_signal, sampling_rate, bits) = audio.wavread(infile)
```

```python
fs = sampling_rate
total_time = len(input_signal)/sampling_rate
frame_size = 0.050 # 50 milliseconds
hop_size = 0.020 # 20 milliseconds
```

```python
stft_data = stft(input_signal, sampling_rate, frame_size, hop_size)
```

```python
istft_data = istft(stft_data, sampling_rate, total_time, hop_size)
```

```python
pylab.figure(figsize=(14,8))
imgplot = pylab.imshow(scipy.absolute(stft_data.T), origin='lower', aspect='auto', interpolation='nearest')
imgplot.set_cmap('spectral')
pylab.xlabel('Time')
pylab.ylabel('Frequency')
pylab.show()
```

```python
T1 = int(0.1 * sampling_rate) # first 0.1 second
t = scipy.linspace(0, total_time, total_time*sampling_rate, endpoint=False)
```

```python
plot_large(t[:T1], input_signal[:T1], t[:T1], istft_data[:T1])
pylab.xlabel('Time (seconds)')
```
