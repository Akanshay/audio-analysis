```python
sample_rate = 10 # samples per sec
total_sampling_time = 3

t = linspace(0, total_sampling_time, total_sampling_time * sample_rate)
f = lambda x: 5 * sin(2*pi*x) + 3 * sin(6*pi*x)
    
sampled_f = [f(i) for i in t]
```

```python
plot(t, sampled_f)
```

```python
fft_output = numpy.fft.fft(sampled_f)
```

```python
freqs = numpy.fft.fftfreq(sample_rate * total_sampling_time) * sample_rate
```

```python
mag = [sqrt(i.real**2 + i.imag**2) for i in fft_output]
```

```python
plot(freqs, mag)
```
