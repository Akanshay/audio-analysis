```python
import scipy
from __future__ import division

def plot_large(*args, **kwargs):
    fig = pylab.figure()
    fig.set_size_inches(14,8)
    plot(*args, **kwargs)
```

```python
sample_rate = 100
total_time = 10
t = linspace(0, total_time, total_time * sample_rate)
f = [1 for i in t]
```

```python
plot_large(t,f)
```

```python
window_size = 100 # 100 points (which is 1 second in this case)
hop_size = window_size // 2
```

```python
window = scipy.hamming(window_size)
```

```python
def flatten(lst):
    return [item for sublist in lst for item in sublist]

window_times = [t[i:i+window_size] for i in range(0, len(f)-window_size, hop_size)]
window_graphs = [[wtime, window] for wtime in window_times]
flattened_window_graphs = flatten(window_graphs)
plot_large(*flattened_window_graphs)
```

```python
windowed = [window * f[i:i+window_size] for i in range(0, len(f)-window_size, hop_size)]
```

```python
convoluted = scipy.zeros(total_time * sample_rate)
for n,i in enumerate(range(0, len(f)-window_size, hop_size)):
    convoluted[i:i+window_size] += windowed[n]
```

```python
plot_large(t, f, t, convoluted)
```
